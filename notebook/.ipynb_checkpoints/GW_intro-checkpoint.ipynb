### Background


#### Chirp mass and the Tidal Deformability

Tidal deformability is a measure of distortion in neutron stars due to the tidal field, often represented by $\Lambda$. It's an important parameter to understand the internal structure of a star and the EOS for very dense objects, and it can be measured through the gravitational waves emitted during neutron star mergers. 


Tidal deformability is defined by,

$
\Lambda = \frac{2}{3} k_2 C^{-5}
$

Where, $C = 2GM/Rc^2$ and is the dimensionless compactness parameter, and $k_2$  is the tidal Love number, see $\ref{rybicki1979}$


Chirp mass is the total mass of these two orbiting neutron stars, which can be precisely measured through the gravitational wave signal from a neutron star merger. It's also one of the important parameters to study the properties of merging objects.



Chirp mass is defined by, 

$
\mathcal{M} = \frac{(M_1 M_2)^{3/5}}{(M_1 + M_2)^{1/5}}
$
