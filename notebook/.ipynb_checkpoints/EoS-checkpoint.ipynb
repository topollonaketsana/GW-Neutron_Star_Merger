### Binary system of Neutron stars: 

We derived the expression for distance r (orbital seperation) using the equations of motion

Using orbital dynamics (frequency vs mass‚Äìdistance) to plot approximate neutron star M‚Äìr relation plot


### The orbital distance r between two neutron stars

### $ r = [\frac{GM}{4 \pi^2 f^2} ]^ \frac{1}{3}$

```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from scipy.integrate import solve_ivp
```

#### Assumptions: before using real data
1. Ideal
2. G = 1
3. assuming mass vary from 0.5 ‚Äì 3ùëÄ‚äô
4. Assume ùëì = 1000 Hz

```python
# plot using the assumptions
G = 1        
Msun = 1.98847e30         
pi = np.pi
f = 1000    

# Mass range in solar masses
M_solar = np.linspace(0.5, 3.0, 100)
M_kg = M_solar * Msun

# orbital radius from the formula
r_m = (G * M_kg / (2 * pi * f)**2) ** (1/3)

# Convert to km for plot
r_km = r_m / 1e3

# Plot
plt.plot(M_solar, r_km)
plt.xlabel('Total Mass $M$ ($M_\\odot$)')
plt.ylabel('Distance $r$ (km)')
plt.title('Mass‚ÄìOrbital-Separation curve')


plt.show()
```

### Virial Theorem:


We can also derive the Polytropic Equation of State from the adiabat equation (no heat exchange) `dQ = 0`

####   $PV^\gamma = K $ 

K is constant, and we can express P interms of density $\rho = \frac{m}{V} $, and therefore:

####   $P = K \rho^ \gamma$

```python
rho = np.logspace(14, 18, 100)  # g/cm^3, typical for neutron stars
gamma = 2.0
K = 1e-2                        # constant to scale

P = K * rho**gamma

plt.loglog(rho, P)
plt.xlabel(f'Density $\\rho$ (g/cm$^3$)')
plt.ylabel('Pressure $P$ (dyn/cm$^2$)')
plt.title('Polytropic Equation of State: $P = K \\rho^\\gamma$')

#plt.grid(True, which='both')
plt.show()
```

### Using multiple $\rho$ - Indices

```python
# Polytropic Equation of State: P = K * rho^gamma
# Define density range (in g/cm^3)
rho = np.logspace(14, 18, 500)  # typical neutron star core densities

# Define multiple polytropic indices 
polytropes = {
    'n = 1.5 (gamma = 5/3)': 5/3,
    'n = 1 (gamma = 2)': 2.0,
    'n = 0.5 (gamma = 3)': 3.0
}

# Polytropic constant K (units arbitrary for now)
K = 1e-2


for label, gamma in polytropes.items():
    P = K * rho**gamma
    plt.loglog(rho, P, label= label)

# Plot
plt.xlabel('Density')
plt.ylabel('Pressure')
plt.title('Polytropic Equation of State')

plt.legend()
plt.show()
```
